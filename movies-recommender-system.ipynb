```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
# credits['cast'].head(1).values
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies.head(2)
```

```python
movies.columns
```

```python
required_attributes = ['movie_id','title','overview','genres','keywords','cast','crew']
```

```python
movies = movies[required_attributes]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    result = []
    for ele in ast.literal_eval(obj):
        result.append(ele['name'])
    return result
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.genres.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.keywords.head(1)
```

```python
def convert_cast(obj):
    result = []
    count = 0
    for ele in ast.literal_eval(obj):
        if count==5:
            break
        result.append(ele['name'])
        count += 1
    return result
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['cast'].head(1).values
```

```python
def fetch_director(obj):
    result = []
    for ele in ast.literal_eval(obj):
        if ele['job']=='Director':
            result.append(ele['name'])
    return result
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.overview[0]
```

```python
movies.overview = movies.overview.apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(lambda x: [c.replace(" ","") for c in x])
movies.keywords = movies.keywords.apply(lambda x: [c.replace(" ","") for c in x])
movies.cast = movies.cast.apply(lambda x: [c.replace(" ","") for c in x])
movies.crew = movies.crew.apply(lambda x: [c.replace(" ","") for c in x])
```

```python
movies.head(3)
```

```python
movies['tags'] = movies.overview + movies.genres + movies.keywords + movies.cast + movies.crew
```

```python
movies.tags.head(1)
```

```python
final_data = movies[['movie_id','title','tags']]
```

```python
final_data.head()
```

```python
final_data.tags = final_data.tags.apply(lambda x: " ".join(x))
```

```python
final_data.tags[1]
```

```python
final_data.tags = final_data.tags.apply(lambda x: x.lower())
```

```python
final_data.tags[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer(max_features=5000,stop_words='english')
```

```python
text_vector = count_vec.fit_transform(final_data['tags']).toarray()
```

```python
text_vector
```

```python
text_vector.shape
```

```python
count_vec.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
por_stem = PorterStemmer()
```

```python
def stem(text):
    result = []
    for c in text.split():
        result.append(por_stem.stem(c))
    return " ".join(result)
```

```python
final_data.tags = final_data.tags.apply(stem)
```

```python
final_data.tags[0]
```

```python
text_vector = count_vec.fit_transform(final_data['tags']).toarray()
text_vector
```

```python
count_vec.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_mat = cosine_similarity(text_vector)
```

```python
sim_mat
```

```python
def recommender(movie_name):
    movie_index = final_data[final_data.title == movie_name].index[0]
    distance = sim_mat[movie_index]
    movie_list = sorted(list(enumerate(distance)),key=lambda tup: tup[1],reverse=True)[1:6]

    recommended_movies = []
    for movie in movie_list:
        movie_id = movie[0]
        recommended_movies.append(final_data.iloc[movie[0]].title)
    
    return recommended_movies
```

```python
recommender('Spectre')
```
